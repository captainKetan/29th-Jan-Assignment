# Answer 1

Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Here are five common plots that you can create using Seaborn, along with their typical uses:

1. **Scatter Plot:**
   - **Use:** Scatter plots are used to visualize the relationship between two continuous variables. Each data point is represented as a point on the graph, making it suitable for displaying patterns, correlations, and outliers in data.
   - **Seaborn Function:** "sns.scatterplot()"

2. **Histogram:**
   - **Use:** Histograms are used to display the distribution of a single continuous variable. They divide the data into bins and represent the frequency or density of data points within each bin.
   - **Seaborn Function:** "sns.histplot()"

3. **Bar Plot:**
   - **Use:** Bar plots are used to visualize categorical data by displaying the values of different categories as bars. They are useful for comparing the values of different categories or showing how a categorical variable varies across different groups.
   - **Seaborn Function:** "sns.barplot()"

4. **Box Plot (Box-and-Whisker Plot):**
   - **Use:** Box plots are used to visualize the distribution of a continuous variable or the comparison of distributions between multiple categories or groups. They display summary statistics such as quartiles, medians, and potential outliers.
   - **Seaborn Function:** "sns.boxplot()"

5. **Heatmap:**
   - **Use:** Heatmaps are used to visualize relationships and patterns in data by displaying a matrix of values as a grid of colored cells. They are particularly useful for showing correlations between variables or exploring patterns in a dataset.
   - **Seaborn Function:** "sns.heatmap()"

# Answer 2

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset=sns.load_dataset('fmri')
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.lineplot(x='timepoint', y='signal', data=dataset, hue='region')
plt.subplot(1,2,2)
sns.lineplot(x='timepoint', y='signal', data=dataset, hue='event')
plt.tight_layout(pad=5)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset=sns.load_dataset('titanic')
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(x='pclass', y='age', data=dataset)
plt.figure(figsize=(10,10))
plt.subplot(1,2,2)
sns.boxplot(x='pclass', y='fare', data=dataset)
plt.tight_layout(pad=5)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data = sns.load_dataset("flights")
data_pivot = data.pivot_table(index="month", columns="year", values="passengers")
plt.figure(figsize=(10, 6))
sns.heatmap(data_pivot, annot=True, fmt="d", cmap="YlGnBu", linewidths=.5)
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Count Heatmap")
plt.show()
```
