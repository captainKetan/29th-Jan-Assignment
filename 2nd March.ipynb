# Answer 2

A **scatter plot** is a type of data visualization that displays individual data points on a two-dimensional (2D) graph. Each data point is represented as a single point on the plot, with its position determined by the values of two variables, one plotted along the x-axis and the other along the y-axis. Scatter plots are commonly used to explore and visualize the relationship or correlation between two continuous variables.

Key characteristics of scatter plots include:

1. **X and Y Axes:** Scatter plots have two axes: the horizontal x-axis and the vertical y-axis. Each axis represents a different variable, allowing you to examine how changes in one variable relate to changes in the other.

2. **Individual Data Points:** Each data point in your dataset is represented as a single point on the scatter plot. The position of each point is determined by the values of the corresponding variables.

3. **Markers or Dots:** Data points are typically represented as markers or dots on the plot. These markers can vary in shape, size, or color, making it possible to distinguish between different categories or groups within the data.

4. **No Lines Connecting Points:** Unlike line plots, scatter plots do not connect data points with lines. This absence of lines emphasizes the individual data points and their distribution.

5. **Correlation Analysis:** Scatter plots are often used to visually assess the correlation or relationship between the two variables. Depending on the shape of the scatter plot, you can infer whether there is a positive, negative, or no correlation between the variables.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('This is a Scatter plot')
plt.show()
```

# Answer 3

The "subplot()" function in Matplotlib is used to create a grid of subplots within a single figure or canvas. It allows you to display multiple plots or visualizations in a structured grid format, making it easier to compare and analyze different aspects of your data. The primary purposes and benefits of using the "subplot()" function include:

1. **Multiple Plots in One Figure:** With "subplot()", you can create and arrange multiple plots within a single figure. This is useful when you want to display related visualizations or compare different datasets side by side.

2. **Custom Layouts:** You have control over the layout of the subplots, including the number of rows and columns in the grid. This flexibility allows you to design complex arrangements of subplots tailored to your specific needs.

3. **Consistent Figure Size:** Subplots ensure that all visualizations within a figure share the same dimensions and are neatly aligned, providing a consistent and professional appearance.

4. **Easy Comparison:** When working with related data or multiple datasets, subplots make it easy to compare trends, patterns, or distributions across different parts of the same figure.

5. **Efficient Use of Space:** Subplots help optimize the use of space within your figure. Instead of creating separate figures for each plot, you can arrange them compactly in a grid, saving space and making your visualizations more accessible.

6. **Common Axes Labels and Titles:** Subplots can share common axis labels (e.g., x-axis and y-axis labels) and titles, which is useful for scenarios where the subplots represent variations or subsets of the same data.

```python
import matplotlib.pyplot as plt
import numpy as np
plt.subplot(2,2,1)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x,y)

plt.subplot(2,2,2)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x,y)

plt.subplot(2,2,3)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x,y)

plt.subplot(2,2,4)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x,y)
```

# Answer 4

A **bar plot**, also known as a **bar chart**, is a graphical representation of data using rectangular bars or columns. Each bar's length or height is proportional to the value it represents, making bar plots ideal for visualizing categorical data or comparing discrete categories. Bar plots are commonly used for the following purposes:

1. **Comparing Categories:** Bar plots are used to compare different categories or groups based on a specific metric or value. For example, you can compare sales figures for different products, the performance of various teams, or the frequency of specific events.

2. **Displaying Frequency Distributions:** Bar plots are effective for showing the distribution or frequency of categories within a dataset. For example, you can use a bar plot to display the frequency of different age groups in a survey.

3. **Visualizing Rankings:** Bar plots are useful for ranking categories based on a particular criterion. For instance, you can use a bar chart to display the top-selling products in a store.

4. **Tracking Changes Over Time:** When bar plots are stacked or grouped, they can be used to track changes in categories over time. This is common in cases like stock price movements over months or sales trends across years.

Key characteristics of bar plots include:

- **Vertical or Horizontal Bars:** Bars can be oriented either vertically (the default) or horizontally, depending on the data and your preference. Vertical bars are typically used when categories have long names, while horizontal bars may be preferable for a small number of categories with short names.

- **Spacing Between Bars:** The bars are separated by a small gap or space to distinguish one category from another. The width or thickness of the bars can be adjusted based on the data and visual preferences.

- **Axes:** Bar plots have two axes: a categorical axis (usually the x-axis) and a numerical axis (usually the y-axis). The categorical axis represents the categories, while the numerical axis represents the values or counts associated with each category.

- **Colors and Labels:** Bar plots often use different colors for each category or group to make them visually distinct. Labels are added to the bars to indicate the exact values they represent.

Bar plots are an essential tool in data visualization because they provide a straightforward and intuitive way to represent and compare categorical data. They are widely used in various fields, including business, finance, science, and social sciences, for data exploration, presentation, and analysis.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.bar(company, profit)
plt.title('Bar plot')
plt.subplot(1,2,2)
plt.barh(company, profit)
plt.title('Horizontal bar plot')
plt.tight_layout(pad=5.0)
plt.show()
```

A **box plot**, also known as a **box-and-whisker plot**, is a graphical representation of the distribution of a dataset. It is commonly used to visualize the central tendency, spread, and skewness of data, especially when dealing with continuous or numerical data. Box plots are particularly useful for the following purposes:

1. **Summary of Data Distribution:** Box plots provide a concise summary of the data distribution by displaying key statistics such as the median, quartiles (25th and 75th percentiles), and potential outliers.

2. **Identification of Outliers:** Box plots help in identifying potential outliers or extreme values in a dataset. Outliers are typically represented as individual points beyond the "whiskers" of the plot.

3. **Comparison of Distributions:** Box plots can be used to compare the distributions of multiple datasets side by side. This makes it easy to visualize differences in central tendency, spread, and skewness between groups.

Key components of a box plot include:

- **Box:** The central box in the plot represents the interquartile range (IQR), which contains the middle 50% of the data. The top and bottom edges of the box represent the 75th percentile (Q3) and the 25th percentile (Q1), respectively. The length of the box indicates the spread or variability of the data within this middle range.

- **Median (Q2):** A vertical line inside the box represents the median, which is the middle value of the dataset when it is ordered. It divides the data into two equal halves.

- **Whiskers:** The "whiskers" extend from the edges of the box to the minimum and maximum values within a defined range. The length of the whiskers can vary depending on the specific method used to calculate them.

- **Outliers:** Outliers are data points that fall outside the range defined by the whiskers. They are often plotted as individual points beyond the whiskers to draw attention to extreme values.

Box plots are widely used in various fields, including statistics, data analysis, and data visualization, for their ability to effectively summarize and compare datasets. They are particularly valuable when dealing with skewed data or when comparing multiple groups or categories to gain insights into data distribution and variation.

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.subplot(1,2,1)
plt.boxplot(box1)
plt.subplot(1,2,2)
plt.boxplot(box2)
plt.tight_layout(pad=5)
plt.show()
```
